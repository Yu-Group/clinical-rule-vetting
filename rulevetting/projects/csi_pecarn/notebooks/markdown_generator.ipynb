```python
data_dictionary = pd.DataFrame(index=eda_df.columns)
data_dictionary.index = data_dictionary.index.rename('Covariate Name')
```

```python
bool_cols = [col for col in eda_df if 
               eda_df[col].dropna().value_counts().index.isin([0,1]).all()]

cond_prob = [eda_df[eda_df[col]==1]['csi_injury'].mean() for col in bool_cols]
```

```python
cond_prob
```

```python
data_dictionary['Type'] = eda_df.dtypes

check_binary = lambda col: np.isin(col.dropna().unique(), [0, 1]).all()
data_dictionary['Binary'] = eda_df.apply(check_binary,axis=0)

#data_dictionary['% Missing'] = pd.isna(eda_df).mean(axis=0) * 100

data_dictionary['Max'] = np.nan
data_dictionary['Max'][data_dictionary.loc[:,'Type'].copy()==np.float64] = eda_df.loc[:,data_dictionary.loc[:,'Type'].copy()==np.float64].max()

data_dictionary['Min'] = np.nan
data_dictionary['Min'][data_dictionary.loc[:,'Type'].copy()==np.float64] = eda_df.loc[:,data_dictionary.loc[:,'Type'].copy()==np.float64].min()

data_dictionary['Mean'] = np.nan
data_dictionary['Mean'][data_dictionary.loc[:,'Type'].copy()==np.float64] = eda_df.loc[:,data_dictionary.loc[:,'Type'].copy()==np.float64].mean()

data_dictionary['Median'] = np.nan
data_dictionary['Median'][data_dictionary.loc[:,'Type'].copy()==np.float64] = eda_df.loc[:,data_dictionary.loc[:,'Type'].copy()==np.float64].median()

data_dictionary['CSI Probability if Indicated'] = np.nan
data_dictionary.loc[bool_cols,'CSI Probability if Indicated'] = cond_prob

```
